```python
np.random.seed(1)
K = 2
N_pts = 20
shap_vals_all = []
n_iters = []
top_K = []

for i in range(N_pts):
    converged = False
    while not converged:
        xloc = X_test[i:(i+1)]
        shap_vals, diffs, converged = rankshap(model, X_train, xloc, K=K, alpha=0.2, 
                                mapping_dict=mapping_dict, max_n_perms=10000, n_init=100, n_equal=False)
        if converged:
            break
    est_top_K = get_ranking(shap_vals)[:K]
    top_K.append(est_top_K)
    avg_length = np.mean([len(diffs[j]) for j in range(d)])
    n_iters.append(avg_length)
    if len(n_iters) % 2 == 0:
        print(len(n_iters), np.mean(n_iters))
```
